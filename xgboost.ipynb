```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split , GridSearchCV 
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import accuracy_score ,classification_report , confusion_matrix 
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error, r2_score

```

```python
df = pd.read_csv("/home/teddy/Music/Java/dataset1/house_price_regression_dataset.csv")
```

```python
x = df.drop('House_Price',axis=1)
y= df['House_Price']
```

```python
x
```

```python
y
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.2,random_state=35)
```

```python
xgb_regression = xgb.XGBRegressor()
```

```python
xgb_regression.fit(x_train,y_train)
```

```python
xgb_regression.predict(x_test)
```

```python
mean_squared_error(y_test,xgb_regression.predict(x_test) )
```

```python
mean_absolute_error(y_test,xgb_regression.predict(x_test))
```

```python
r2_score(y_test,xgb_regression.predict(x_test))
```

```python
param_grid={
    "n_estimators":[50,100,200],
    "learning_rate":[.01,.1,.2],
    "max_depth":[3,5,7,8,9,11],
    "subsample":[.7,.8,1],
    "colsample_bytree":[.7,.8,.1]
}
```

```python
grid_search =GridSearchCV(xgb.XGBRegressor(),param_grid,cv=3,n_jobs=1,verbose=2)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
xgb_regression_best_params = xgb.XGBRegressor(**grid_search.best_params_)
```

```python
xgb_regression_best_params.fit(x_train,y_train)
```

```python
xgb_regression_best_params.predict(x_test)
```

```python
mean_squared_error(y_test,xgb_regression_best_params.predict(x_test) )
```

```python
mean_absolute_error(y_test,xgb_regression_best_params.predict(x_test))
```

```python
r2_score(y_test,xgb_regression_best_params.predict(x_test))
```
